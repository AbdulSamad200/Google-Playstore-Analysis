# 3.0 Data Preprocessing
Dropping unnecessary columns and handling missing values are common preprocessing steps performed before conducting Exploratory Data Analysis (EDA) for several reasons:

1. **Reducing Dimensionality:** Dropping unnecessary columns helps reduce the dimensionality of the dataset, making it easier to focus on the most relevant features for analysis. This simplification can lead to more efficient computation and visualization during EDA.

2. **Improving Interpretability:** Removing irrelevant or redundant columns improves the interpretability of the analysis results. By focusing on meaningful features, you can extract clearer insights and draw more accurate conclusions about the data.

3. **Enhancing Data Quality:** Handling missing values ensures that the data used for analysis is of high quality. Missing values can introduce bias and inaccuracies into the analysis if not properly addressed. By dropping rows or imputing missing values, you can ensure the integrity and reliability of the analysis results.

4. **Facilitating Visualization:** EDA often involves creating visualizations to explore the data distribution, relationships, and patterns. Dropping unnecessary columns and handling missing values can make it easier to create informative and visually appealing plots, enhancing the understanding of the data.

5. **Streamlining Analysis:** Removing unnecessary columns and missing values streamlines the analysis process by focusing on the most relevant data points. This can save time and computational resources, especially when working with large datasets.

### Question 2: What is the distribution of app sizes across each category

```python
def convert_size_to_numeric(size):
    if pd.isnull(size) or size == 'Varies with device':
        return None
    size = str(size).replace(',', '')  
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1000 
    else:
        return None

df['Size_numeric'] = df['Size'].apply(convert_size_to_numeric)

# Removing Outliers  using IQR (Inter Quartile Range)
def remove_outliers_iqr(df, column, category_column):
    grouped = df.groupby(category_column)
    filtered_data = pd.DataFrame()
    for group_name, group_data in grouped:
        Q1 = group_data[column].quantile(0.25)
        Q3 = group_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        filtered_group = group_data[(group_data[column] >= lower_bound) & (group_data[column] <= upper_bound)]
        filtered_data = pd.concat([filtered_data, filtered_group])
    return filtered_data

# Removing outliers from 'Size_numeric' column within each category
df_filtered = remove_outliers_iqr(df, 'Size_numeric', 'Category')

# Plotting the distribution of app sizes within each category after removing outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Size_numeric', data=df_filtered)
plt.xticks(rotation=90) 
plt.title('Distribution of App Sizes by Category (After Removing Outliers)')
plt.xlabel('Category')
plt.ylabel('App Size (MB)')
plt.show()
```

### Question 6: What is the trend release in app over time? 

```python
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')

# Extract year and month from the 'Released' column
df['Year'] = df['Released'].dt.year
df['Month'] = df['Released'].dt.month

# Group by year and month to count the number of app releases
app_releases = df.groupby(['Year', 'Month']).size().reset_index(name='Count')

# Plot the trend of app releases over time
plt.figure(figsize=(12, 6))
plt.plot(app_releases['Year'], app_releases['Count'], marker='o', linestyle='-')
plt.title('Trend of App Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of App Releases')
plt.grid(True)
plt.show()
```

### Question 7: Are there any patterns in the update frequency  of the app?

```python
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')

# Calculate the time difference between 'Last Updated' and 'Released' dates
df['Update Frequency'] = (df['Last Updated'] - df['Released']).dt.days

# Plot the distribution of update frequencies
plt.figure(figsize=(10, 6))
df['Update Frequency'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of App Update Frequencies')
plt.xlabel('Update Frequency (Days)')
plt.ylabel('Number of Apps')
plt.grid(True)
plt.show()
```
